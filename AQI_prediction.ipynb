```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaled_df=scaler.fit_transform(new_df)
```

```python
scaled_df=pd.DataFrame(scaled_df,columns=new_df.columns)
type(scaled_df)
```

```python
scaled_df.describe()
```

```python
y=scaled_df.iloc[:,0]
X=scaled_df.iloc[:,1:]
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regressor=LinearRegression()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
linear_regressor.fit(X_train,y_train)
```

```python
y_train_pred=linear_regressor.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.metrics import r2_score
```

```python
train_error=mean_squared_error(y_train,y_train_pred)
train_score=r2_score(y_train,y_train_pred)
```

```python
print(train_error)
print(train_score)
```

```python
y_test_pred=linear_regressor.predict(X_test)
```

```python
test_error=mean_squared_error(y_test,y_test_pred)
test_score=r2_score(y_test,y_test_pred)
```

```python
print("Test Error =",test_error)
print("Test Score =",test_score)
```

```python
prediction=linear_regressor.predict([[0.9,2.6,22.6,27.2]])       # Real time data and AQI value for this is 27
print(prediction)
```

```python
plt.figure(figsize=(15,10))
plt.plot(np.array(y_test[:100]))
plt.plot(y_test_pred[:100])
plt.legend(['Actual','Predicted'])
plt.title('First 100 samples')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
grboost=GradientBoostingRegressor()
grboost.fit(X_train,y_train)
```

```python
y_gbr_pred=grboost.predict(X_test)
```

```python
mse_gbr=mean_squared_error(y_test,y_gbr_pred)
r2_gbr=r2_score(y_test,y_gbr_pred)
print("Mean Squared Error of Gradient boosting =",mse_gbr)
print("R2score of Gradient Boosting =",r2_gbr)
```

```python
plt.figure(figsize=(15,10))
plt.plot(np.array(y_test[:100]))
plt.plot(y_gbr_pred[:100])
plt.legend(['Actual','Predicted'])
plt.title('First 100 samples')
```

```python
plt.subplot(1,2,1)

plt.plot(np.array(y_test[:100]))
plt.plot(y_test_pred[:100])
plt.legend(['Actual','Predicted'])
plt.title("Linear Regression vs Gradient Boosting Regression")
plt.subplot(1,2,2)

plt.plot(np.array(y_test[:100]))
plt.plot(y_gbr_pred[:100])
```
